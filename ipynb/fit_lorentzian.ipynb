```python
f, s = ana.get_spectra(sig, line_broadening=2, zerofill=100, filt_method=None)
f_ppm = utils.freq_to_ppm(f)
idx = utils.ppm_idx(f_ppm, 2.6, 3.4)
```

```python
xx = [plot(s[x,1][idx] + s[x,0][idx]) for x in range(s.shape[0])]
```

```python
# For two_lorentzian: 
# freq0_1, freq0_2, area1, area2, hwhm1, hwhm2, phase1, phase2, offset, drift
initial = 2.9, 3.2, 0.1, 0.1, 0.01, 0.01, 0, 0, 0, 0
n_params = 10
fit_func = utils.two_lorentzian
bounds = [(2.6,3.6),(2.6, 3.6),(0,None),(0,None),(0,None),(0,None),(-np.pi, np.pi),(-np.pi, np.pi),(None,None),(None,None)]

# For a single lorentzian: 
# freq0, area, hwhm, phase, ffset, drift
initial = 3.0, 0.1, 0.1, 0, 0, 0
n_params = 6
fit_func = utils.lorentzian
bounds = [(2.6,3.6),(0,None),(0,None),(-np.pi, np.pi),(None,None), (None, None)]

```

```python
model = np.empty((s.shape[0], idx.stop - idx.start))
signal = np.empty((s.shape[0], idx.stop - idx.start), dtype=np.complex64)
params = np.empty((s.shape[0], n_params))
for ii, xx in enumerate(s):
    signal[ii] = xx[1][idx]
    params[ii], _ = lsq.leastsqbound(mopt.err_func, initial, args=(f_ppm[idx],  np.real(signal[ii]), fit_func), bounds=bounds)
    model[ii] = fit_func(f_ppm[idx], *params[ii])
```
